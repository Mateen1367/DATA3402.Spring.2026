*Exercise 1:* Write a function that creates an n by n matrix (of list of lists) which will represent the state of a Tie Tac Toe game. Let 0, 1, and 2 represent empty, "X", and "O", respectively.


````python
# Write your solution here
def tictactoe(n):
    """
    Creates an n x n Tic-Tac-Toe board.
    0 = empty, 1 = X, 2 = O
    """
    board = []
    for _ in range(n):
        board.append([0]*n)
    return board
````

*Exercise 4:* Write a function that takes a `n` by `n` matrix representing a tic-tac-toe game, and returns -1, 0, 1, or 2 indicating the game is incomplete, the game is a draw, player 1 has won, or player 2 has one, respectively. Here are some example inputs you can use to test your code:

````python
def game_status(board):
    """
    Returns:
    -1 : game incomplete
     0 : draw
     1 : player 1 wins (X)
     2 : player 2 wins (O)
    """
    n = len(board)

    # check rows
    for r in range(n):
        if board[r][0] != 0 and all(board[r][c] == board[r][0] for c in range(n)):
            return board[r][0]

    # check columns
    for c in range(n):
        if board[0][c] != 0 and all(board[r][c] == board[0][c] for r in range(n)):
            return board[0][c]

    # check main diagonal
    if board[0][0] != 0 and all(board[i][i] == board[0][0] for i in range(n)):
        return board[0][0]

    # check other diagonal
    if board[0][n-1] != 0 and all(board[i][n-1-i] == board[0][n-1] for i in range(n)):
        return board[0][n-1]

    # check if incomplete (any zeros left)
    if any(0 in row for row in board):
        return -1

    # otherwise draw
    return 0
````

````python
winner_is_2 = [[2, 2, 0],
    [2, 1, 0],
    [2, 1, 1]]

winner_is_1 = [[1, 2, 0],
    [2, 1, 0],
    [2, 1, 1]]

winner_is_also_1 = [[0, 1, 0],
    [2, 1, 0],
    [2, 1, 1]]

no_winner = [[1, 2, 0],
    [2, 1, 0],
    [2, 1, 2]]

also_no_winner = [[1, 2, 0],
    [2, 1, 0],
    [2, 1, 0]]

tests = {
    "winner_is_2": winner_is_2,
    "winner_is_1": winner_is_1,
    "winner_is_also_1": winner_is_also_1,
    "no_winner": no_winner,
    "also_no_winner": also_no_winner
}

for name, b in tests.items():
    print(name, "->", game_status(b))
````

*Exercise 5:* Write a function that takes a game board, player number, and `(x,y)` coordinates and places "X" or "O" in the correct location of the game board. Make sure that you only allow filling previously empty locations. Return `True` or `False` to indicate successful placement of "X" or "O".

````python
# Write your solution here
def place_move(board, player, x, y):
    # check bounds
    if x < 0 or y < 0 or x >= len(board) or y >= len(board):
        return False

    # only place if empty
    if board[x][y] == 0:
        board[x][y] = player
        return True

    return False
````

````python
# Test your solution here
board = [
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 0]
]

print(place_move(board, 1, 0, 0))  # True
print(place_move(board, 2, 0, 0))  # False (already filled)
print(place_move(board, 2, 1, 1))  # True

for row in board:
    print(row)
````

*Exercise 9:* Use all of the previous exercises to implement a full tic-tac-toe game, where an appropriate board is drawn, 2 players are repeatedly asked for a location coordinates of where they wish to place a mark, and the game status is checked until a player wins or a draw occurs.

*Exercise 10:* Test that your game works for 5x5 Tic Tac Toe.  

````python
# Test your solution here
def play_tictactoe(n):
    # create empty board
    board = [[0]*n for _ in range(n)]
    player = 1

    def draw(board):
        print("   " + "  ".join(str(i+1) for i in range(n)))
        print("  " + " ---"*n)
        for i in range(n):
            row = chr(ord('A')+i) + " "
            for j in range(n):
                if board[i][j] == 1:
                    row += "| X "
                elif board[i][j] == 2:
                    row += "| O "
                else:
                    row += "|   "
            row += "|"
            print(row)
            print("  " + " ---"*n)

    def status(board):
        # rows & columns
        for i in range(n):
            if board[i][0] != 0 and all(board[i][j] == board[i][0] for j in range(n)):
                return board[i][0]
            if board[0][i] != 0 and all(board[j][i] == board[0][i] for j in range(n)):
                return board[0][i]

        # diagonals
        if board[0][0] != 0 and all(board[i][i] == board[0][0] for i in range(n)):
            return board[0][0]
        if board[0][n-1] != 0 and all(board[i][n-1-i] == board[0][n-1] for i in range(n)):
            return board[0][n-1]

        if any(0 in row for row in board):
            return -1
        return 0

    while True:
        draw(board)
        move = input(f"Player {player} move (e.g. A1): ").upper()
        r = ord(move[0]) - ord('A')
        c = int(move[1]) - 1

        if 0 <= r < n and 0 <= c < n and board[r][c] == 0:
            board[r][c] = player
        else:
            print("Invalid move, try again.")
            continue

        result = status(board)
        if result == 1:
            draw(board)
            print("Player 1 (X) wins!")
            break
        elif result == 2:
            draw(board)
            print("Player 2 (O) wins!")
            break
        elif result == 0:
            draw(board)
            print("Game is a draw.")
            break

        player = 2 if player == 1 else 1
````

````python
play_tictactoe(5)
````

*Exercise 11:* Develop a version of the game where one player is the computer. Do not do an extensive search for the best move. Make computer player can simply protect against loosing and otherwise try to win with straight or diagonal patterns.

````python
# Write you solution here
def play_vs_computer(n):
    board = [[0]*n for _ in range(n)]

    def draw():
        print("   " + "  ".join(str(i+1) for i in range(n)))
        print("  " + " ---"*n)
        for i in range(n):
            row = chr(ord('A')+i) + " "
            for j in range(n):
                row += "| X " if board[i][j]==1 else "| O " if board[i][j]==2 else "|   "
            print(row+"|")
            print("  " + " ---"*n)

    def status():
        for i in range(n):
            if board[i][0]!=0 and all(board[i][j]==board[i][0] for j in range(n)):
                return board[i][0]
            if board[0][i]!=0 and all(board[j][i]==board[0][i] for j in range(n)):
                return board[0][i]
        if board[0][0]!=0 and all(board[i][i]==board[0][0] for i in range(n)):
            return board[0][0]
        if board[0][n-1]!=0 and all(board[i][n-1-i]==board[0][n-1] for i in range(n)):
            return board[0][n-1]
        if any(0 in row for row in board):
            return -1
        return 0

    def computer_move():
        # try to win or block
        for player in [2, 1]:
            for i in range(n):
                for j in range(n):
                    if board[i][j] == 0:
                        board[i][j] = player
                        if status() == player:
                            board[i][j] = 2
                            return
                        board[i][j] = 0
        # otherwise first empty
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = 2
                    return

    while True:
        draw()
        move = input("Your move (e.g. A1): ").upper()
        r = ord(move[0]) - ord('A')
        c = int(move[1]) - 1

        if 0 <= r < n and 0 <= c < n and board[r][c] == 0:
            board[r][c] = 1
        else:
            print("Invalid move")
            continue

        if status() == 1:
            draw(); print("You win!")
            break

        computer_move()

        if status() == 2:
            draw(); print("Computer wins!")
            break
        if status() == 0:
            draw(); print("Draw!")
            break
````

````python
# Test your solution here
play_vs_computer(3)
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
# Write you solution here
def play_vs_computer_minimax(n, max_depth=4):
    board = [[0]*n for _ in range(n)]

    def draw():
        print("  " + " ---"*n)
        for i in range(n):
            row = ""
            for j in range(n):
                row += "| X " if board[i][j]==1 else "| O " if board[i][j]==2 else "|   "
            print(row + "|")
            print("  " + " ---"*n)

    def status():
        for i in range(n):
            if board[i][0]!=0 and all(board[i][j]==board[i][0] for j in range(n)):
                return board[i][0]
            if board[0][i]!=0 and all(board[j][i]==board[0][i] for j in range(n)):
                return board[0][i]
        if board[0][0]!=0 and all(board[i][i]==board[0][0] for i in range(n)):
            return board[0][0]
        if board[0][n-1]!=0 and all(board[i][n-1-i]==board[0][n-1] for i in range(n)):
            return board[0][n-1]
        if any(0 in r for r in board):
            return -1
        return 0

    def minimax(depth, maximizing):
        s = status()
        if s != -1 or depth == 0:
            return 1 if s==2 else -1 if s==1 else 0

        best = -999 if maximizing else 999
        for i in range(n):
            for j in range(n):
                if board[i][j]==0:
                    board[i][j] = 2 if maximizing else 1
                    val = minimax(depth-1, not maximizing)
                    board[i][j] = 0
                    best = max(best,val) if maximizing else min(best,val)
        return best

    def computer_move():
        best, move = -999, None
        for i in range(n):
            for j in range(n):
                if board[i][j]==0:
                    board[i][j]=2
                    val = minimax(max_depth-1, False)
                    board[i][j]=0
                    if val > best:
                        best, move = val, (i,j)
        board[move[0]][move[1]] = 2

    while True:
        draw()
        move = input("Your move (row col): ").split()
        r,c = int(move[0]), int(move[1])
        if board[r][c]!=0: continue
        board[r][c]=1
        if status()==1: draw(); print("You win"); break
        computer_move()
        if status()==2: draw(); print("Computer wins"); break
        if status()==0: draw(); print("Draw"); break
````

````python
# Test your solution here
play_vs_computer_minimax(3, max_depth=4)
````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. The "smarter" player's win rate for each grid.

````python
# Write you solution here
import random

def simulate_games(n, games=10, max_depth=3):
    smart_wins = 0

    def empty_board():
        return [[0]*n for _ in range(n)]

    def status(b):
        for i in range(n):
            if b[i][0]!=0 and all(b[i][j]==b[i][0] for j in range(n)):
                return b[i][0]
            if b[0][i]!=0 and all(b[j][i]==b[0][i] for j in range(n)):
                return b[0][i]
        if b[0][0]!=0 and all(b[i][i]==b[0][0] for i in range(n)):
            return b[0][0]
        if b[0][n-1]!=0 and all(b[i][n-1-i]==b[0][n-1] for i in range(n)):
            return b[0][n-1]
        if any(0 in r for r in b):
            return -1
        return 0

    def minimax(b, depth, maxp):
        s = status(b)
        if s != -1 or depth == 0:
            return 1 if s==2 else -1 if s==1 else 0
        best = -999 if maxp else 999
        for i in range(n):
            for j in range(n):
                if b[i][j]==0:
                    b[i][j] = 2 if maxp else 1
                    v = minimax(b, depth-1, not maxp)
                    b[i][j] = 0
                    best = max(best,v) if maxp else min(best,v)
        return best

    def smart_move(b):
        best, move = -999, None
        for i in range(n):
            for j in range(n):
                if b[i][j]==0:
                    b[i][j]=2
                    v = minimax(b, max_depth-1, False)
                    b[i][j]=0
                    if v > best:
                        best, move = v, (i,j)
        b[move[0]][move[1]] = 2

    def random_move(b, player):
        moves = [(i,j) for i in range(n) for j in range(n) if b[i][j]==0]
        i,j = random.choice(moves)
        b[i][j] = player

    for _ in range(games):
        b = empty_board()
        while True:
            random_move(b,1)
            if status(b)==1 or status(b)==0: break
            smart_move(b)
            if status(b)==2:
                smart_wins += 1
                break
            if status(b)==0: break

    return smart_wins / games
````

````python
# Test your solution here
for size in [3, 4, 5]:
    rate = simulate_games(size, games=10, max_depth=3)
    print(f"{size}x{size} board: Smart player win rate = {rate}")
````
